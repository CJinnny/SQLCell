```python
%%sql DB=sqlcell NOTIFY=1 MAKE_GLOBAL=df
select distinct '' as blank, '' as blank1,  h.dateof, h.team, h.opp, h.pts, h.fg, h.fg_att, h.ft, h.ft_att, 
    h.fg3, h.fg3_att, h.off_rebounds, h.def_rebounds, h.off_rebounds + h.def_rebounds as totalrebounds,
    h.asst, h.blks, h.fouls, h.stls, h.turnovers,
    a.pts as opoints, a.fg as ofg, a.fg_att as ofgatt, a.ft as oft, a.ft_att as oftatt, 
    a.fg3 as ofg3, a.fg3_att as ofg3att, a.off_rebounds as ooffrebounds, a.def_rebounds as odefrebounds,
    a.off_rebounds + a.def_rebounds as ototalrebounds, a.asst as oassists, a.blks as oblocks, a.fouls as ofouls,
    a.stls as osteals, a.turnovers as oturnovers
from nba h
join nba a 
    on h.opp = a.team
    and h.team = a.opp
    and (h.team <> 'EAS' and a.team <> 'EAS')
    and (h.opp <> 'WES' and a.opp <> 'WES')
where h.dateof = a.dateof
    and h.dateof >= '2016-10-24'
order by h.dateof desc
```
